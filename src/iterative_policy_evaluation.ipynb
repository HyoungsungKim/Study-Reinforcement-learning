```python
import numpy as np
import math
```

```python
grid = np.zeros((20,4,4))
```

```python
ds = [[-1,0], [1, 0], [0,-1], [0,1]]
```

```python
def iterative_policy_evaluation(k, current_state):
    next_state = []
    (x, y) = (current_state[0], current_state[1])
    next_value = 0
    
    for i in range(0,4):        
        state = [x + ds[i][0], y + ds[i][1]]
                
        if (state[0] < 0 or state[0] > 3):
            state[0] = x
        if (state[1] < 0 or state[1] > 3):
            state[1] = y
            
        next_value += (-1 + grid[k-1][state[0]][state[1]])        
    
    grid[k][x][y] = round(next_value/4, 1)    
```

```python
for k in range(1,len(grid)):
    for i in range(0, 4):
        for j in range(0, 4):
            if(i == 0 and j == 0):
                continue
            elif(i == 3 and j == 3):
                continue

            iterative_policy_evaluation(k, [i,j])
```
